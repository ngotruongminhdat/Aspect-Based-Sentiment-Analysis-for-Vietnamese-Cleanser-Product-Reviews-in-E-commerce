```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/home/dat/Desktop/Aspect-Based-Sentiment-Analysis-for-Vietnamese-Cleanser-Product-Reviews-in-E-commerce/Dataset/labeled_data.csv')
data.head(10)
```

```python
df = data.copy()
```

```python
df['QUALITY'] = np.nan
df['SMELL'] = np.nan
df['SHIPPING'] = np.nan
df['PACKING'] = np.nan
df['PRICE'] = np.nan
df['OTHERS'] = np.nan
```

```python
for i in range(0, len(df)):
    items = [x.strip() for x in df.Polarity.iloc[i].split(',')]
    for item in items:
        parts = item.split(':')
        if len(parts) == 2:
            key, value = parts
            df[key].iloc[i] = value
```

```python
df.drop(['Rating', 'Aspect', 'Polarity'], axis = 1, inplace = True)
df.to_csv('/home/dat/Desktop/Aspect-Based-Sentiment-Analysis-for-Vietnamese-Cleanser-Product-Reviews-in-E-commerce/Dataset/data.csv')
```
